```python
import numpy as np
import matplotlib.pyplot as plt
import gefera as gf
```

```python
t = np.linspace(-0.5, 0.2, 10000)

ap = 1.0
tp = -91.25
ep = 0.0
pp = 365
wp = 0.1 * np.pi / 180
ip = 90 * np.pi / 180

am = 0.02
tm = -4.25
em = 0.0
pm = 8
om = 45 * np.pi / 180
wm = -90 * np.pi / 180
im = 90.0 * np.pi / 180
mm = 0.01

u1 = 0.2
u2 = 0.3
rp = 0.2
rm = 0.1
```

```python
po = gf.PrimaryOrbit(ap, tp, ep, pp, wp, ip)
mo = gf.SatelliteOrbit(am, tm, em, pm, om, wm, im, mm)
sys = gf.HierarchicalSystem(po, mo)

fig = plt.figure(figsize=(10, 10))
gf.draw(sys, fig.gca(), -0.35, rp, rm, ld_params=(u1, u2))

plt.savefig('/Users/tgordon/Desktop/titleslide.pdf')
```
